#### Removing StopWords

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
import re
from unidecode import unidecode
import spacy
nlp = spacy.load('en_core_web_sm')

#bad_chars  = [',','']
def preprocessing_text(paragraph):
    
    para_clean = unidecode(paragraph)
    lower_case = para_clean.lower()
    no_numbers = re.sub(r'\d', '', lower_case) #Remove Numbers out of the Text
    no_special_chars = re.sub(r'[^ a-zA-Z]', '', no_numbers) #Remove Special Characters
    #out of the text
    
    return no_special_chars

def stopwords_lemma(paragraph):
    
    empty_par = []
    words_mod_lemma = nltk.tokenize.word_tokenize(paragraph)
    
    #Is taking too much time
    #doc = nlp(paragraph)    
    #words_mod_lemma = [word.lemma_ for word in doc if word not in set(stop_words)]
    words_mod_lemma_2 = [word for word in words_mod_lemma if word not in stop_words]
    
    paragraph_2 = ' '.join(words_mod_lemma_2)
    
    return paragraph_2
```

```python
data_train.title = data_train.title.apply(preprocessing_text)
data_train.text = data_train.text.apply(preprocessing_text)

data_train.title = data_train.title.apply(stopwords_lemma)
data_train.text = data_train.text.apply(stopwords_lemma)
```

```python
data_train
```

### Generate the model analysis

Remember that count vectorizer is called bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, ngram_range= (1,3))
X = cv.fit_transform(data_train.title).toarray()
```

```python
title_count = pd.DataFrame(X, columns = cv.get_feature_names())
```

```python
title_count
```

### Generating Train Test Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, data_train['label'], test_size = 0.3, random_state = 42)
```

### Multinomial NB Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```

```python
from sklearn import metrics
```

```python
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
```

```python
print(metrics.classification_report(y_test, pred))
```

```python
print(metrics.confusion_matrix(y_test, pred))
```

```python
feature_names = cv.get_feature_names()
```

```python
classifier.coef_[0]
```

```python
#This are the words that doesn´t make post to be fake
sorted(zip(classifier.coef_[0], feature_names), reverse = True)[:20]
```

```python
#This are the words that make a post to be fake
sorted(zip(classifier.coef_[0], feature_names))[:20]
```
